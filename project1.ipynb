```python
#glucose
dataset['Glucose']=dataset['Glucose'].replace(0, dataset['Glucose'].median())
#BP
dataset['Glucose']=dataset['Glucose'].replace(0, dataset['Glucose'].median())
#Insulin
dataset['Insulin']=dataset['Insulin'].replace(0, dataset['Insulin'].mean())
#Skin thickness
dataset['SkinThickness']=dataset['SkinThickness'].replace(0, dataset['SkinThickness'].mean())
#BMI
dataset['BMI']=dataset['BMI'].replace(0, dataset['BMI'].mean())
dataset
```

```python
X=dataset.drop(['Outcome'],axis=1)
Y=dataset['Outcome']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy=[]
test_accuracy=[]
for n in range (1,11):

 knn=KNeighborsClassifier(n_neighbors=n)
 knn.fit(X_train,y_train)
 training_accuracy.append(knn.score(X_train,y_train))
 test_accuracy.append(knn.score(X_test,y_test))   #knn.score returns us the mean accuracy

print(training_accuracy)
print(test_accuracy)
```

```python
plt.plot(range(1,11),training_accuracy,label='training accuracy')
plt.plot(range(1,11),test_accuracy,label='test accuracy')
plt.xlabel('Accuracy')
plt.ylabel('n_neigbours')
plt.legend() #this statement helps us to print up the topic of each graph which is being depicted by it
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=0,max_depth=6)
dt.fit(X_train,y_train)
print(dt.score(X_test,y_test))
print(dt.score(X_train,y_train))
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(random_state=42)
mlp.fit(X_train,y_train)
print(mlp.score(X_test,y_test))
print(mlp.score(X_train,y_train))
```
